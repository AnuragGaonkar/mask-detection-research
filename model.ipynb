```python
import tensorflow as tf
import os

# 1. Load the model using Keras 3
model = tf.keras.models.load_model('best_model.keras', compile=False)

# 2. Save it as a standard TensorFlow SavedModel directory
# This "freezes" the Keras 3 specific attributes into standard TF format
temp_saved_model_dir = "temp_saved_model"
model.export(temp_saved_model_dir) 

# 3. Convert from the SavedModel directory
converter = tf.lite.TFLiteConverter.from_saved_model(temp_saved_model_dir)

# 4. Apply optimizations for "Low-End Devices" (from your paper)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.target_spec.supported_types = [tf.float16] # Reduces size by 50% with minimal accuracy loss

tflite_model = converter.convert()

# 5. Save the final file
with open('mask_model_mobile.tflite', 'wb') as f:
    f.write(tflite_model)

print("✅ Success! Optimized TFLite model saved as mask_model_mobile.tflite")
```
